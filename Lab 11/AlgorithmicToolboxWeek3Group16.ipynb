#Semana 3: Greedy Algorithms

##Money Change

```python
a = int(input())
contar = 0
for i in [10, 5, 1]:
    if a>=i:
        q = a//i
        contar = contar + q
        a = a%i
        if a == 0:
            print(contar)
            quit()
```

##Maximum value of loot

```python
n, W = list(map(int, input().split()))
lista = []
vl = 0

if W == 0:
    print(0)
    quit()

for _ in range(n):
    v, w = list(map(int, input().split()))
    if v == 0:
        continue
    lista.append((v, w))

lista.sort(key = lambda x: x[0]/x[1], reverse = True)

for v,w in lista:
    if W==0:
        print(vl)
        quit()
    a = min(w, W)
    vl = vl + a*v/w
    W = W - a

print(vl)
```

##Collecting signatures

```python
n = int(input())
lista = []

for e in range(n):
    a, b = list(map(int, input().split()))
    lista.append((a,b))

lista.sort(key = lambda x: x[1])

i = 0
coord = []

while i < n:
    c = lista[i]
    while i < n-1 and c[1]>=lista[i+1][0]:
        i += 1

    coord.append(c[1])
    i += 1

print(len(coord))

print(' '.join([str(i) for i in coord]))
```
