Función bubble sort optimizada

```python
def BSOpt(arr): 
  # recorre los elementos del arrreglo
  for i in range(len(arr)):
    perm = False
    pasos = 0
    # compara elementos del arreglo
    for j in range(0, len(arr) - i - 1):
      pasos +=1
      if arr[j] > arr[j + 1]:
        #intercambia elementos si no están en orden
        temp = arr[j]
        arr[j] = arr[j+1]
        arr[j+1] = temp
        pasos +=1
        perm = True
    if not perm:
      break
  return pasos
```
