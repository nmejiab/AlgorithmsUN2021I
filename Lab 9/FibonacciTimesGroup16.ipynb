##**Lab 9 - Fibonacci Times**

```python
import matplotlib.pyplot as plt
from time import process_time
```

###**Fibonacci Recursivo**

```python
def fibRecursiva(n):
    if n==0 :
        return 0
    if n==1 :
        return 1
    return fibRecursiva(n-1)+fibRecursiva(n-2)
```

Gráfica del tiempo de ejecución de Fibonacci Recursivo hasta n = 50

```python
n = 50
n_listaR = []
TiempoR = []
for i in range (n):
  n_listaR.append(i)
  inicio = process_time()
  fibRecursiva(i)
  fin = process_time() 
  TiempoR.append(fin-inicio)
plt.title("Tiempos de ejecución Fibo")
plt.plot(n_listaR, TiempoR, label="Fibo Recursivo", color='green')
plt.xlabel("n")
plt.ylabel("Tiempo")
plt.show()
```

Comparación con y = n

```python
plt.title("Comparación con y = n")
plt.plot(n_listaR, TiempoR, label="Fibo Recursivo", color='green')
plt.plot(n_listaR, n_listaR, label="y=n", color='red')
plt.xlabel("n")
plt.ylabel("Tiempo")
plt.show()
```

###**Fibonacci Iterativo Lineal**

```python
def fibIterativoLin(n):
    if n==0 :
        return 0
    a, b = 0, 1
    for i in range(1, n):
        a, b = b, a+b
    return b
```

Gráfica del tiempo de ejecución de Fibonacci Iterativo Lineal hasta n = 1000

```python
n = 1000
n_listaItLin = []
TiempoItLin = []
for i in range (n):
  n_listaItLin.append(i)
  inicio = process_time()
  fibIterativoLin(i)
  fin = process_time() 
  TiempoItLin.append(fin-inicio)
plt.title("Tiempos de ejecución Fibo")
plt.plot(n_listaItLin, TiempoItLin, label="Fibo Iterativo Lineal", color='orange')
plt.xlabel("n")
plt.ylabel("Tiempo")
plt.show()
```

Comparación con y = n

```python
plt.title("Tiempos de ejecución Fibo")
plt.plot(n_listaItLin, TiempoItLin, label="Fibo Iterativo Lineal", color='orange')
plt.plot(n_listaItLin, n_listaItLin, label="y=n", color='red')
plt.xlabel("n")
plt.ylabel("Tiempo")
plt.legend()
plt.show()
```

###**Fibonacci Iterativo Logarítmico**

```python
def fibIterativoLog(n):

    if n==0: return 0  
    v1, v2, v3 = 1, 1, 0  

    for rec in bin(n)[3:]:
        calc = (v2*v2) 
        v1, v2, v3 = (v1*v1+calc) , ((v1+v3)*v2) , (calc+v3*v3) 
        if rec == '1': v1, v2, v3 = (v1+v2), v1, v2
    return v2;
```

Gráfica del tiempo de ejecución de Fibonacci Iterativo Logarítmico hasta n = 1000

```python
n = 1000
n_listaItLog = []
TiempoItLog = []
for i in range (n):
  n_listaItLog.append(i)
  inicio = process_time()
  fibIterativoLog(i)
  fin = process_time() 
  TiempoItLog.append(fin-inicio)
plt.title("Tiempos de ejecución Fibo")
plt.plot(n_listaItLog, TiempoItLog, label="Fibo Iterativo Logarítmico", color='blue')
plt.xlabel("n")
plt.ylabel("Tiempo")
plt.show()
```

Comparación con y = n

```python
plt.title("Tiempos de ejecución Fibo")
plt.plot(n_listaItLog, TiempoItLog, label="Fibo Iterativo Logarítmico", color='blue')
plt.plot(n_listaItLog, n_listaItLog, label="y=n", color='red')
plt.xlabel("n")
plt.ylabel("Tiempo")
plt.legend()
plt.show()
```
