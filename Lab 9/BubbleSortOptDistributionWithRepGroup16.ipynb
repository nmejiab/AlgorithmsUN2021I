##**Lab 9 - Bubble Sort Optimized Distribution With Repetition**

Funci칩n bubble sort optimizada

```python
def BSOpt(arr): 
  # recorre los elementos del arrreglo
  for i in range(len(arr)):
    perm = False
    pasos = 0
    # compara elementos del arreglo
    for j in range(0, len(arr) - i - 1):
      pasos +=1
      if arr[j] > arr[j + 1]:
        #intercambia elementos si no est치n en orden
        temp = arr[j]
        arr[j] = arr[j+1]
        arr[j+1] = temp
        pasos +=1
        perm = True
    if not perm:
      break
  return pasos
```

Permutaciones con repeticiones

```python
from itertools import product

n=3 
perm = product(range(1,n+1), repeat = n)

c=0
for i in perm:
    print(i)
    c+=1

print(c)
```

Funci칩n de histogramas y an치lisis



```python
from itertools import product
import matplotlib.pyplot as plt
import numpy as np 

n = 3
perm = product(range(1,n+1), repeat = n)

times =  [] 
# Print the obtained permutations
for i in list(perm):
    i = list(i)
    times.append(BSOpt(i))

b = np.array(range(min(times)-1,max(times)+2) ) +0.5
plt.hist(times,bins=b)
plt.title("Bubble Sort Optimized repetitions")
plt.xlabel("Steps = comparisons + swaps ")
plt.ylabel("Frequency")
plt.show()
    
print("n",n)  
print("number of permutations ,n**n = ",n**n)
print("exp min =",min(times))
print("min formula (n-1)*n/2 =",(n-1)*n/2)
print("exp max =",max(times))
print("max formula (n-1)*n =",(n-1)*n)
print("exp mean =",np.mean(times))
print("mean formula (3/4)*(n**2)-n+(1/4) =",(3/4)*(n**2)-n+(1/4))
print("stdev",np.std(times))
```
