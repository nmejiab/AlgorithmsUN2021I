##Maximum amount of gold

```python
P = list(map(int, input().split()))
z = list(map(int, input().split()))
a = list(map(int, input().split()))
v = [[0]*(len(a)+1) for _ in range(P+1)]
for i in range(1,len(a)+1):
    for j in range(1,P+1):
        v[j][i] = v[j][i-1]
        if a[i-1] <= j:
            valor = v[j-a[i-1]][i-1] + a[i-1]
            if valor >= v[j][i]:
                v[j][i] = valor
print(v[P][len(a)])
```

##Partitioning souvenirs

```python
import numpy as np
x = int(input())
w = list(map(int, input().split()))
au = sum(w)

if x<3: 
    print('0')
elif au % 3 != 0:
    print('0')
else:
    p = au//3
    ct = 0 
    val = np.zeros((p+1, x+1))
    for i in range(1, p+1):
        for j in range(1, x+1):
            val[i][j] = val[i][j-1]
            if w[j-1]<=i:
                tp = val[i-w[j-1]][j-1] + w[j-1]
                if tp > val[i][j]:
                    val[i][j] = tp
            if val[i][j] == p:
                ct += 1
    if ct < 3:
        print('0')
    else:
        print('1')
```
