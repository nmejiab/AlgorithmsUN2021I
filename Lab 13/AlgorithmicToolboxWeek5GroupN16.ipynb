#Semana 5: Dynamic programming

##Money change again

```python
import sys
din = int(input())
ca = [1, 3, 4]
mon_m = [0] + [sys.maxsize]*din

for i in range(1, din+1):
    for j in ca:
        if i>=j:
            cambio = mon_m[i-j]+1
            if cambio < mon_m[i]:
                mon_m[i] = cambio
print(mon_m[din])
```

##Primitive Calculator

```python
import math
n = int(input())
op = [0, 0] + [math.inf]*(n-1)
for i in range(2, n+1):
    t_1, t_2, t_3 = [math.inf]*3
    t_1 = op[i-1] + 1 
    if i%2 == 0: t_2 = op[i//2] + 1
    if i%3 == 0: t_3 = op[i//3] + 1
    mop = min(t_1, t_2, t_3)
    op[i] = mop
print(op[n])

nms = [n]
while n!=1:
    if n%3 ==0 and op[n]-1 == op[n//3]:
        nms += [n//3]
        n = n//3
    elif n%2 ==0 and op[n]-1 == op[n//2]:
        nms += [n//2]
        n = n//2
    else:
        nms += [n-1]
        n = n - 1

print(' '.join([str(i) for i in nms][::-1]))
```

#Longest Common Subsequence of Two Sequences

```python
x = int(input())
a = list(map(int, input().split()))
y = int(input())
b = list(map(int, input().split()))

lcs = [[0 for i in range(y + 1)] for j in range(x + 1)]
for i in range(1, x+1):
    for j in range(1, y+1):
        if a[i-1] == b[j-1]:
            lcs[i][j] = lcs[i-1][j-1] + 1
        elif a[i-1] != b[j-1]:
            lcs[i][j] = max(lcs[i-1][j-1], lcs[i-1][j], lcs[i][j-1])
print(lcs[x][y])
```
